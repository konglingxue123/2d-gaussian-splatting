```python
import os
import cv2
import numpy as np

mask_dir = '/home/ubunto/Project/konglx/pcd/2dgs/2d-gaussian-splatting-main/datasets/dalian_xinghaiwandaqiao_video_input_rgba/outputs/SegmentationClass'
img_dir = '/home/ubunto/Project/konglx/pcd/2dgs/2d-gaussian-splatting-main/datasets/dalian_xinghaiwandaqiao_video_input_rgba/outputs/JPEGImages'
save_dir = '/home/ubunto/Project/konglx/pcd/2dgs/2d-gaussian-splatting-main/datasets/dalian_xinghaiwandaqiao_video_input_rgba/images_rgba'
os.makedirs(save_dir, exist_ok=True)

from PIL import Image

# def trans_rgba_to_rgb_black_bg_new(image_path, save_path=None):
#     try:
#         imagePtah = image_path
#         img = Image.open(imagePtah)
#         # if img.mode != 'RGBA':
#         #     image = img.convert('RGBA')
#         width = img.width
#         height = img.height

#         image = Image.new('RGBA', size=(width, height), color=(0, 0, 0, 0))
#         image.paste(img, (0, 0), mask=img)
#         if save_path:
#             image.save(save_path)
#         # image.show()
#         print(f'image {imagePtah.split("/")[-1]} has been processed.')
#         return image
        
#     except Exception as e:
#         print(e)
n = 0

for img_name in os.listdir(img_dir):
    n += 1
    img_name = img_name.split('.')[0]
    img_each_dir = os.path.join(img_dir, img_name+'.jpg')
    # img_rgba = trans_rgba_to_rgb_black_bg(img_each_dir, os.path.join(save_dir, img_name+'.png'))
    img_pil = Image.open(os.path.join(img_dir, img_name+'.jpg'))
    
    img_pil_np = np.array(img_pil)
    r, g, b = cv2.split(img_pil_np)
    
    mask_pil = Image.open(os.path.join(mask_dir, img_name+'.png'))
    # mask_pil_rgba = mask_pil.convert('RGBA')
    mask_pil_np = np.array(mask_pil)
    mask_pil_np = (mask_pil_np * 255).astype(np.uint8)
    
    img = cv2.merge((r, g, b, mask_pil_np))
    img_pil = Image.fromarray(img.astype(np.uint8))

    img_pil.save(os.path.join(save_dir, img_name+'.png'))
    
    # print(img_pil_rgba_np.shape, np.max(img_pil_rgba_np), mask_pil_np.shape, np.unique(mask_pil_np))
    # print('##')
    print(f'num: {n}, image {img_name} has been processed.')
```
